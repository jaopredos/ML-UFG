## Relação com a variável alvo

```python
df_ana = df.copy()
df_ana.head()
```

```python
def analyze_categorical(df, col):
    if col not in df.columns: return
    print(f"--- Relação: {col} vs Transported ---")
    # Agrupa por valor, conta total e calcula média de transportados
    summary = df.groupby(col)['Transported'].agg(['count', 'mean']).sort_values(by='mean', ascending=False)
    summary['mean'] = summary['mean'].round(4)
    print(summary)
    print("\n")

# Colunas Categóricas Básicas
cat_cols = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']
for col in cat_cols:
    analyze_categorical(df_ana, col)

# Engenharia de 'Cabin' (Deck e Lado)
# A coluna Cabin é 'Deck/Num/Side' (ex: B/0/P)
if 'Cabin' in df_ana.columns:
    # Cria colunas temporárias
    df_ana['Deck'] = df_ana['Cabin'].str.split('/', expand=True)[0]
    df_ana['Side'] = df_ana['Cabin'].str.split('/', expand=True)[2]
    
    analyze_categorical(df_ana, 'Deck')
    analyze_categorical(df_ana, 'Side')

# Análise de Gastos (Zero vs Maior que Zero)
spend_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for col in spend_cols:
    col_spent = f'Spent_{col}'
    df_ana[col_spent] = df_ana[col] > 0
    analyze_categorical(df_ana, col_spent)
```
